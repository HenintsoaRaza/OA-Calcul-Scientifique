```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import math
```

### 1. Instanciation des paramètres du modèle

```python
# Maturité:
T = 1

# Le strike vaut:
K = 100

# le taux de la banque centrale européene
r = 0.1

# la volatilité:
sigma = 0.3

# temps:
t = np.arange(0, 1, 1/180)

# s:
s = range(1, 301)
```

### 2. Calcul de $d_1$ et $d_2$

```python
d1 = np.zeros((len(s), len(t))) 
d2 = np.zeros((len(s), len(t))) 
for i in range(len(t)):
    for j in range(len(s)):
        d1[j][i] = (np.log(s[j]/K) + (r + 0.5 * (sigma)**2)*(T-t[i])) / (sigma * np.sqrt(T-t[i]))
        d2[j][i] = (np.log(s[j]/K) + (r - 0.5 * (sigma)**2)*(T-t[i])) / (sigma * np.sqrt(T-t[i]))
d1
```

### 3. Calcul du Call $C(t,S)$

```python
C = np.zeros((len(s), len(t))) 
for i in range(len(t)):
    for j in range(len(s)):
        a = s[j] * 0.5 * math.erfc(-d1[j][i] / np.sqrt(2))
        b = K*np.exp(-r*(T-t[i]))*(0.5*math.erfc(-d2[j][i] / np.sqrt(2)))
        C[j][i] = a - b
        
print(C.shape)
```

### 4. Création des points de coordonnées $(t , S , C(t,S))$ pour les représentations graphiques

```python
n = 0
N = len(s)*len(t)
coord = np.zeros((N, 3)) 
for i in range(len(t)):
    for j in range(len(s)):
        coord[n][2] = C[j][i]
        coord[n][0] = j
        coord[n][1] = i
        n += 1
coord
```

### 5. Représentation graphique 2D de la valeur du Call $C(S)$ pour $t=0$

```python
plt.figure()

plt.plot(s, C[:, 0])

plt.xlabel('S')
plt.ylabel('Valeur du Call')
plt.title('Valeur du Call pour t = 0 ')
plt.show()
```

### 6. Représentation graphique 3D de la valeur du Call $C(t,S)$

```python
colors = cm.rainbow(np.linspace(0, 1, 100))

# Color according to the value of C
col_C = []
nb_col = colors.shape[0]
Cmax = max(coord[:,2])
for c in coord[:,2]:
    col_index = int((c/Cmax) * nb_col) - 1
    if col_index < 0:
        col_index = 0
    col_C.append(colors[col_index])
    
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(coord[:, 1], coord[:, 0] , coord[:, 2], c=col_C, marker='o')

ax.set_xlabel('t')
ax.set_ylabel('S')
ax.set_zlabel('C')

plt.gca().invert_xaxis()

plt.show()
```
